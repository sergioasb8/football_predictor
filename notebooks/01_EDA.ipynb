```python
def plot_pattern(data:pd.DataFrame, position:int, y_label:str, width=15, height=6):

    '''
    Info: 
        Function to create a line plot to identify patterns across time based on the final position by a team at the end of the season.
    ------
    Input:
        * data: data to generate the plot (type : pd.DataFrame)
        * position: final team position (type : int)
        * y_label: column name that will be the reference to create lineplot (type : str)
        * width: width for figure size (type : int)
        * height: height for figure size (type : int)
    ------
    Output:
        * line plot to reflect patterns accross time
    '''

    # figure size in inches
    sns.set(rc={'figure.figsize':(width,height)})

    for season in data['season'].unique().tolist():
        # find the team who finish 7th at the end of the season
        team = data[data['season']==season].drop_duplicates(subset=['Team'], keep='last').query(f"position == {position}")['Team'].reset_index(drop=True)

        # get the data for the team that finished 7th
        team_season_data = data.query(f"season == '{season}'").query(f"Team == {team[0]}").reset_index(drop=True).reset_index().rename(columns={'index':'match'})
        team_season_data['match'] = team_season_data['match']+1
        
        # sns.lineplot(data=team_season_data, x='match', y=y_label,label=f'{season}')
        sns.lineplot(data=team_season_data, x='match', y=y_label)
```

## 2.4 Position patterns

Looking for pattern according to the position.

```python
plot_pattern(data,1,'position')
```

```python
plot_pattern(data,15,'position')
```

```python
plot_pattern(data,20,'position')
```
